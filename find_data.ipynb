```python
import ee
import geemap
ee.Initialize()
```

```python
Map = geemap.Map(center=[70,90], zoom=4)
Map
```

```python
Map.add_basemap('SATELLITE') #Google Satellite Hybrid;Google Maps;Google Terrain;Google Satellite;OpenStreetMap.Mapnik
```

```python
dataset = ee.ImageCollection('NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG').filter(ee.Filter.date('2017-05-01', '2017-05-31'))
nighttime = dataset.select('avg_rad')
nighttimeVis = {min: 0.0, max: 60.0};
# Map.setCenter(-77.1056, 38.8904, 8);
Map.addLayer(nighttime, nighttimeVis, 'Nighttime')
```
